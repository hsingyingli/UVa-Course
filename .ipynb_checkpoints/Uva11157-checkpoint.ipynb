```python
path = "/home/lab/Downloads/11157.in"
fp = open(path,"r")
```

```python
n_cases = int(fp.readline())
cases = 0
while  n_cases:
    cases = cases +1
    buff = fp.readline().split(" ")
    N,D = buff[0] , buff[1]
    buff = fp.readline()
    frog1 = 0
    frog2 = 0
    distance = []
    end = " B-" + D
    buff = buff+end
    info = buff.replace("\n","").split(" ")
    while (frog1 != int(D) or frog2 != int(D)):
        for i in range(len(info)):
            if(int(info[i][2:]) > frog1):
                distance.append(int(info[i][2:].replace("\n","")) - frog1)
                frog1 = int(info[i][2:])
                if(info[i][0:1] == 'S'):
                    info.remove(info[i])
                break
        for i in range(len(info)):
            if(int(info[i][2:]) > frog2):
                distance.append(int(info[i][2:].replace("\n","")) - frog2)
                frog2 = int(info[i][2:])
                if(info[i][0:1] == 'S'):
                    info.remove(info[i])
                break
    print("Case %d: %d"%(cases,max(distance)))
    n_cases = n_cases-1
```
